```python
def display_menu():
    print ('Welcome to the Employee Management program')
    print ()
    print ('COMMAND MENU:')
    print ('show - show all employees')
    print ('add - add an employee')
    print ('del - delete an employee')
    print ('exit - exit program')
    print ()
    
    
def display(current_list):
    for i in range(len(current_list)):
        print(i+1, current_list[i], end="\n") 
    print()


def add(current_list):
    toAdd= input('Enter the emplyee name: ')
    current_list.append(toAdd.strip())
    print(toAdd.strip(), 'was added. \n')

def delete(current_list):
    toDelete = input('Enter the number of the person to delete from the list: ')
    if int(toDelete)>0 and int(toDelete)<len(current_list):
        current_list=current_list.pop(int(toDelete)-1)
    else:
        print("Wrong number")


def main():
    employee_list = ['Zak Ruvalcaba', 'Sally Smith', 'Mark Jones', 'John Jacobson']
    display_menu()
    while True:
        command = input ('Enter command: ')
        if command.lower()=='show':
            display(employee_list)
        elif command.lower() == 'add':
            add(employee_list)
        elif command.lower() == 'del':
            delete(employee_list)
        elif command.lower() == 'exit':
            break
        else:
            print ('That is not a valid command. Please try again. \n')
    print("Program has been terminated")
if __name__=='__main__':
    main()
```

```python
def display_menu():
    print ('Welcome to the Employee Management program')
    print ()
    print ('COMMAND MENU:')
    print ('show - show all employees')
    print ('add - add an employee')
    print ('del - delete an employee')
    print ('exit - exit program')
    print ()
    
    
def display(current_list):
    if (len(current_list)>0):
        for i in range(len(current_list)):
            print(f"{i+1}. Name: {current_list[i][0]}, \nTitle: {current_list[i][1]}, \nYears in company: {current_list[i][2]}\n") 
        print()
    else:
        print("No employees in the list")


def add(current_list):
    toAddName= input('Enter the emplyee name: ')
    toAddTitle= input('Enter the emplyee title: ')
    toAddYears= input('Enter the years in a company: ')
    toAdd = [toAddName, toAddTitle, int(toAddYears)]
    current_list.append(toAdd)
    print(toAdd[0], 'was added. \n')

def delete(current_list):
    toDelete = input('Enter the number of the person to delete from the list: ')
    if int(toDelete)>0 and int(toDelete)<=len(current_list):
        print(current_list[int(toDelete)-1][0], 'will be deleted')
        current_list=current_list.pop(int(toDelete)-1)
        
    else:
        print("Wrong number")


def main():
    employee_list = [['Zak Ruvalcaba', 'CEO', 10] ,
                     ['Sally Smith', 'Manager', 5],
                     ['Mark Jones', 'Web Developer', 6],
                     ['John Jacobson', 'QA', 8]]
    display_menu()
    while True:
        command = input ('Enter command: ')
        if command.lower()=='show':
            display(employee_list)
        elif command.lower() == 'add':
            add(employee_list)
        elif command.lower() == 'del':
            delete(employee_list)
        elif command.lower() == 'exit':
            break
        else:
            print ('That is not a valid command. Please try again. \n')
    print("Program has been terminated")
if __name__=='__main__':
    main()
```
